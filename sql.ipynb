```python
from sqlalchemy import create_engine
import pandas as pd

engine = create_engine('postgresql://postgres:1234@localhost/CovidPortfolio')
```

```python
df = pd.read_csv('CovidDeaths.csv', sep = ';')
df.head()
```

```python
df.to_sql('CovidDeaths', engine, if_exists='replace')
```

```python
df2 = pd.read_csv('CovidVaccs.csv', sep = ';')
df2.head()
```

```python
df2.to_sql('CovidVaccinations', engine, if_exists='replace')
```
